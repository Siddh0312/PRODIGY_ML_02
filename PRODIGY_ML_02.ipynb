```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('/content/Mall_Customers.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
```

```python
km = KMeans(n_clusters = 5)
y_pred = km.fit_predict(df[['Annual Income (k$)', 'Spending Score (1-100)']])
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
df3 = df[df['cluster'] == 3]
df4 = df[df['cluster'] == 4]

plt.scatter(df0['Annual Income (k$)'], df0['Spending Score (1-100)'], color = 'green')
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color = 'red')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color = 'blue')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color = 'yellow')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color = 'black')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
df3 = df[df['cluster'] == 3]
df4 = df[df['cluster'] == 4]

plt.scatter(df0['Annual Income (k$)'], df0['Spending Score (1-100)'], color = 'green')
plt.scatter(df1['Annual Income (k$)'], df1['Spending Score (1-100)'], color = 'red')
plt.scatter(df2['Annual Income (k$)'], df2['Spending Score (1-100)'], color = 'blue')
plt.scatter(df3['Annual Income (k$)'], df3['Spending Score (1-100)'], color = 'yellow')
plt.scatter(df4['Annual Income (k$)'], df4['Spending Score (1-100)'], color = 'black')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color = 'purple', marker = '*', label = 'centroid')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
k_rng = range(1,100)
sse = []
for k in k_rng:
  km = KMeans(n_clusters = k)
  km.fit(df[['Annual Income (k$)', 'Spending Score (1-100)']])
  sse.append(km.inertia_)
sse[99]
```
